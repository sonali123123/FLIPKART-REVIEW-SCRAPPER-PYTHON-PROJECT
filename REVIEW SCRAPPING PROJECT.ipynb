```python
flipkart_url = "https://www.flipkart.com/search?q="+"iphone11"
```

```python
flipkart_url
```

```python
response_website = urReq(flipkart_url)  #Through this URL Request, we are able to hit this URL using Python
```

```python
data_flipcart = response_website.read()  #This information is same as View Page Source of Flipkart page written in HTML
```

```python
beautified_html = bs(data_flipcart,"html.parser")  #Some indentation change: Beautifying information
```

```python
beautified_html  #Now we have all the information on Flipkart Iphone 11 page
```
